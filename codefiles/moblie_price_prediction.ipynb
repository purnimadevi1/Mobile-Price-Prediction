```python
from preprocessing import load_and_preprocessor_data
from prediction_analysis import plot_prediction_accuracy_per_class
from decision_boundary import plot_decision_boundary_with_pca
from sklearn.metrics import confusion_matrix,accuracy_score
```

# **Preprocessing**

```python
pipeline,X_train,y_train,X_test,y_test=load_and_preprocessor_data(r'C:\Users\Purni\Documents\ML Project\datasets\train.csv')
```

# **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_model = LogisticRegression(penalty= 'l1',solver ='saga',random_state=0)
lr_model.fit(X_train,y_train)
```

```python
y_pred = lr_model.predict(X_test)
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

```python
plot_prediction_accuracy_per_class(y_test,y_pred)
```

# **K Nearest Neighbours**

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
## 3, 7, 11, 15
knn_model = KNeighborsClassifier(n_neighbors = 15,weights = 'distance')
knn_model.fit(X_train,y_train)
```

```python
y_pred = knn_model.predict(X_test)
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

```python
plot_prediction_accuracy_per_class(y_test,y_pred)
```
